# Travel Dataset TBank

### Датасет с информацией об онлайн-платежах и клиентах Т-Путешествия

Содержит данные о **заказах** (типы, статусы), **демографические характеристики** клиентов (пол, возраст, образование) и их **поведенческие метрики** (первые/последние заказы, каналы коммуникации).


```python
import pandas as pd

df = pd.read_csv('dano_dataset_travel.csv', sep=None, engine='python')

df.head()
```

### Посмотрим основные показатели датасета

```python
print("Размер датасета (строки, колонки):", df.shape)

print("\nКолонки:")
for col in df.columns:
    print(f"- {col}")

print("\nТипы данных:", df.dtypes.value_counts())

print("\nПропуски:", df.isnull().sum().sum())
```

### Как-то много пропусков, посмотрим где они

```python
print("Пропуски в колонках кол-во/проценты:")
for col in df.columns:
    missing = df[col].isnull().sum()
    if missing > 0:
        percent = round(missing / len(df) * 100, 2)
        print(f"{col:30} : {missing:6} пропусков ({percent:5.2f}%)")
```

### Хочу удалить все столбцы, где пропусков >50% 

```python
missing_percent = df.isnull().sum() / len(df) * 100
columns_to_keep = missing_percent[missing_percent <= 50].index.tolist()
df_clean = df[columns_to_keep].copy()

missing_in_clean = df_clean.isnull().sum() / len(df_clean) * 100
missing_data = missing_in_clean[missing_in_clean > 0].sort_values(ascending=False)

print("Пропуски в очищенном датасете:")
print("=" * 80)
print(f"{'№':4} {'Колонка':50} {'Процент':10}")
print("-" * 80)

for i, (col, percent) in enumerate(missing_data.items(), 1):
    print(f"{i:4} {col:50} {percent:10.6f}%")

print("=" * 80)
print(f"Всего колонок с пропусками: {len(missing_data)}")
print(f"Было: {df.shape}")
print(f"Стало: {df_clean.shape}")
print(f"Удалено колонок: {df.shape[1] - df_clean.shape[1]}")
```

### Поиск дубликатов, проверка адекватности типов данных

```python
# 1. ТИПЫ ДАННЫХ
print("\nТИПЫ ДАННЫХ:")
print(df_clean.dtypes.value_counts())
print("\nДетально:")
print(df_clean.dtypes)

# Проверка на подозрительные типы
print("\n🔍 Подозрительные колонки:")
for col in df_clean.columns:
    if df_clean[col].dtype == 'object':
        # Проверка на даты в object колонках
        sample = df_clean[col].dropna().iloc[0] if len(df_clean[col].dropna())>0 else ''
        if isinstance(sample, str) and any(x in sample for x in ['-', ':', '/']):
            print(f"  ⚠️  {col}: похоже на дату (object -> datetime)")
    elif df_clean[col].dtype in ['int64', 'float64']:
        # Проверка на категориальные числовые
        if df_clean[col].nunique() < 20:
            print(f"  ℹ️  {col}: мало уникальных значений ({df_clean[col].nunique()}) - возможно категория")

# 2. ДУБЛИКАТЫ
print("\n" + "="*60)
print("\nДУБЛИКАТЫ")

# Полные дубликаты
full_dupl = df_clean.duplicated().sum()
print(f"Полных дубликатов строк: {full_dupl}")
if full_dupl > 0:
    print(f"  ({full_dupl/len(df_clean)*100:.2f}% от всех строк)")
    print("\nПримеры дубликатов:")
    print(df_clean[df_clean.duplicated(keep=False)].head())

# Дубликаты по ключевым колонкам
id_cols = [col for col in df_clean.columns if any(x in col.lower() for x in ['id', 'rk', 'key'])]
if id_cols:
    print("\nДубликаты по идентификаторам:")
    for col in id_cols[:3]:  # первые 3
        dupl = df_clean[col].duplicated().sum()
        if dupl > 0:
            print(f"  ⚠️  {col}: {dupl} дубликатов")

# 3. ИТОГ
print("\n" + "="*60)
print("ИТОГ:")
print(f"Строк: {df_clean.shape[0]}, Колонок: {df_clean.shape[1]}")
print(f"Числовых: {len(df_clean.select_dtypes(include=['number']).columns)}")
print(f"Категориальных: {len(df_clean.select_dtypes(include=['object']).columns)}")
print(f"Дубликаты: {'Есть' if full_dupl>0 else 'Нет'}")
```
